# Here we will analyse spread out Bragg peak (SOBP) data from the GEANT4 simulation and compare it to the experimental data.

```python
#Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
import scipy.special as sp
import pandas as pd
```

```python
#reding data from file

data = pd.read_excel('dose_spreadout.ods', header=0)
data.columns = ['Depth','Dose']
data
```

### GEANT4 GENERATED BRAGG PEAK USING O16 WITH 992 MeV energy

```python
plt.plot(data['Depth']/10, data['Dose']/data['Dose'].max(), color = 'black', linewidth = 1)
plt.xlim(0,8)
plt.grid()

plt.xlabel('Depth [mm]')
plt.ylabel('Dose [a.u.]')
plt.title('Bragg Peak using Geant4 generated data')
plt.show()
```

## NOW PLOTTING EXPERIMENTAL DATA


```python
data1 = pd.read_excel('exp_spreadout.ods', header = None)
data1.columns = ['Depth', 'Dose'] 
data1
```

```python
plt.plot(data1['Depth'], data1['Dose']/data1['Dose'].max(), color = 'red', linewidth = 1)
plt.xlim(0,9)
plt.xticks(np.linspace(0,9,300))
plt.xticks(np.arange(0,9,1))
plt.grid()
plt.xlabel('Depth [mm]')
plt.ylabel('Dose [a.u.]')
plt.title('Bragg Peak using experimental data')
plt.show()
```

### Bragg peak for 16O using experimental data obtained with silicon detector

## COMAPRISON OF SIMULATED AND EXPERIMENTAL DATA

```python
plt.plot(data['Depth']/10, data['Dose']/data['Dose'].max(),color = 'black', linewidth = 1 )
plt.scatter(data1['Depth'], data1['Dose']/data1['Dose'].max(),marker='s', facecolors='black', edgecolors='r', s=10 )

plt.xlim(0,9)
plt.grid()
plt.xlabel('Depth [mm]')
plt.ylabel('Dose [a.u.]')
plt.legend(['Geant4', 'Experimental'])
plt.title('Bragg Peak comparison of experimental and Geant4 generated data')


#energy 
plt.show()
```

## Now we are proceeding with LET calculation

### data are obtained from geant4 simulation and can be found in let.out file.

```python
dataLET = pd.read_excel('let_oxygenmac_fin.ods', header = 0)
dataLET
```

## We caluclate separatelly LET track and LET dose

# First we calculate LET dose

```python
dataLET.columns = ['Depth','LET Dose Total','LET track total','O16_D', 'O16_T', 'Let Dose', 'LET track']

plt.scatter(dataLET['Depth']/10, dataLET['LET Dose Total'], color = 'black', marker="3", s=10, linewidth = 1)
plt.scatter(dataLET['Depth']/10, dataLET['Let Dose'], color = 'purple', marker='s', facecolors='none', s=30, linewidth = 1)
plt.xlim(0,7)

plt.grid()

plt.xlabel('Depth [mm]')
plt.ylabel('LET Dose Toatal [MeV/um]')
plt.title('LET Dose total Geant4 generated data')
plt.legend(['LET Dose Total', 'LET Dose Primary'])
plt.show()
```

# Then we calculate LET track

```python
dataLET.columns = ['Depth','LET Dose Total','LET track total','O16_D', 'O16_T', 'Let Dose', 'LET track']

plt.scatter(dataLET['Depth']/10, dataLET['LET track total'], color = 'black', marker="4", s=10, linewidth = 1)
plt.scatter(dataLET['Depth']/10, dataLET['LET track'], color = 'blue', marker='o', facecolors='none', s=30, linewidth = 1)

plt.xlim(0,7)

plt.grid()

plt.xlabel('Depth [mm]')
plt.ylabel('LET Track Toatal [MeV/um]')
plt.title('LET track total Geant4 generated data')
plt.legend(['LET track total', 'LET track Primary'])
plt.show()
```

# Finally on the same plot we have LET dose and Bragg peak obtained from geant4 simulation and from experimental data

```python
import matplotlib.pyplot as plt

# Define the figure and axis objects
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Plot the first data on the left y-axis
ax1.plot(data['Depth']/10, data['Dose']/data['Dose'].max(), color='black', linewidth=1)
ax1.scatter(data1['Depth'], data1['Dose']/data1['Dose'].max(), marker='s', facecolors='black', edgecolors='r', s=10)
ax1.set_xlabel('Depth [mm]')
ax1.set_ylabel('Dose [a.u.]')
ax1.set_xlim(0, 7)
ax1.grid()
ax2.scatter(dataLET['Depth']/10, dataLET['LET Dose Total'], color='black', marker="3", s=10, linewidth=1)
ax2.scatter(dataLET['Depth']/10, dataLET['Let Dose'], color='purple', marker='s', facecolors='none', s=30, linewidth=1)
ax2.set_ylabel('LET Dose [KeV/um]')


#common legend
fig.legend(['Geant4', 'Experimental', 'LET-Total-Dose', 'LET-Dose'], loc='upper left', ncol=4)
plt.show()

```

# Now we plot LET track and Bragg peak obtained from Geant4 simulations and from experimental data on the same plot.

```python
import matplotlib.pyplot as plt

# Define the figure and axis objects
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Plot the first data on the left y-axis
ax1.plot(data['Depth']/10, data['Dose']/data['Dose'].max(), color = 'black', linewidth = 1)
#ax1.plot(data['depth']/10, data['Dose(Gy)']/data['Dose(Gy)'].max(), color = 'black', linewidth = 1)

#ax1.plot(data['Depth']/10, data['Dose']/data['Dose'].max(), color='black', linewidth=1)
ax1.scatter(data1['Depth'], data1['Dose']/data1['Dose'].max(), marker='s', facecolors='black', edgecolors='r', s=10)
ax1.set_xlabel('Depth [mm]')
ax1.set_ylabel('Dose [a.u.]')
ax1.set_xlim(0, 7)
ax1.grid()
#ax1.legend(['Geant4', 'Experimental'])
#ax1.set_title('Bragg Peak comparison of experimental and Geant4 generated data')

# Plot the second data on the right y-axis

plt.scatter(dataLET['Depth']/10, dataLET['LET track total'], color = 'black', marker="4", s=10, linewidth = 1)
plt.scatter(dataLET['Depth']/10, dataLET['LET track'], color = 'blue', marker='o', facecolors='none', s=30, linewidth = 1)
ax2.set_ylabel('LET track [KeV/um]')


#common legend
fig.legend(['Geant4 - Oxygen contribution', 'Experimental', 'LET Track Total', 'LET Track'], loc='upper left', ncol=4)
plt.show()

```

## COMPARISON WITH EXPERIMENTAL RESULTS

### We have calculated yf and yd for LET track and LET dose respectively. Now we are going to compare them.
### yf and yd are calculated using data obtained from MCA. 

```python
#read yf and yd text files
experimental_data = np.loadtxt('yd_yf_data.txt')
#column 0 is yd, column 1 is yf column 2 is depth
yf = experimental_data[:,0]
yd = experimental_data[:,1]
depth = experimental_data[:,2]
```

```python
plt.scatter(depth, yd, marker="d", facecolors='red', edgecolors='red', s=20)


plt.scatter(dataLET['Depth']/10, dataLET['LET Dose Total'], color = 'black', marker='o', s=10, linewidth = 1)
plt.scatter(dataLET['Depth']/10, dataLET['Let Dose'], color = 'blue', marker='s', facecolors='none', s=45, linewidth = 1)
plt.xlim(0,7)
plt.xlabel('Depth [mm]')
plt.ylabel('LET Dose [KeV/um]')
plt.grid()
plt.legend(['yd', 'LET Dose Total', 'LET Dose Primary'])
plt.show()
```

```python
#plot let track total and let track primary and yf
plt.scatter(depth, yf, marker="d", facecolors='red', edgecolors='red', s=30)

plt.scatter(dataLET['Depth']/10, dataLET['LET track total'], color = 'black', marker='o', s=10, linewidth = 1)
plt.scatter(dataLET['Depth']/10, dataLET['LET track'], color = 'blue', marker='s', facecolors='none', s=40, linewidth = 1)  
plt.xlim(0,7)
plt.xlabel('Depth [mm]')
plt.ylabel('LET track [KeV/um]')
plt.grid()
plt.legend(['yf', 'LET track total', 'LET track Primary'])
plt.show()
```
