### This script is used to calculate mean values and errors of yf_exp and yd_exp from csv files obtained 

```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#function to calculate the mean of every column

def mean_values(data):
    for i in data.columns:
        print(i,":",data[i].mean())
```

```python
#make a function to find standard deviation of each column
def std_values(data):
    for i in data.columns:
        print(i, data[i].std())
```

```python
#store mean values and std values in dataframes
def mean_std_yf_yd(data):
    mean_values(data)
    std_values(data)
    mean = pd.DataFrame(data.mean())
    std = pd.DataFrame(data.std())
    mean_std = pd.concat([mean, std], axis=1)
    mean_std.columns = ['Mean of yf', 'standard deviation of yf', 'Mean of yd', 'standard deviation of yd']
    return mean_std_yf_yd
```
